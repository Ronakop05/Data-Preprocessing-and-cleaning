```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data2.csv')
df.head()
```

```python
print(df.describe())
```

```python
df.dropna()
```

```python
avg = ['Age']
data = df[avg].values
avg = np.nanmean(data, axis=0)
print("Average age of the customers is", avg)
```

```python
avg = ['Review Rating']
data = df[avg].values
avg = np.nanmean(data, axis=0)
print("Average Ratings from the customers is", avg)
```

```python
avg = ['Purchase Amount (USD)']
data = df[avg].values
avg = np.nanmean(data, axis=0)
print("Mean Purchase amount of the customers is", avg)
```

```python
max = ['Previous Purchases']
print("Item which maximum purchased",df[max].max())
max_index = df[max].idxmax()
print( df.loc[max_index])
```

```python
plt.hist(df['Age'])
plt.title('Histogram of age')
plt.xlabel('age')
plt.ylabel('orders')
plt.show()
```

this histogram shows that the maximum orders are from age group of 20-25 and 65-70

```python
sns.countplot(x='Item Purchased', data=df, )
plt.title('Items')
plt.ylabel('Count')
plt.xticks(rotation=67)
plt.show()
```

this plot shows that which items are purchased more....

```python
sns.countplot(x='Shipping Type', data=df, )
plt.title('Shipping Type')
plt.ylabel('Count')
plt.xticks(rotation=20)
plt.show()
```

    through this graph we can infer that maximum orders are taken as free shipping as their shipping type

```python
sns.countplot(x='Payment Method', data=df, )
plt.title('Payment Method')
plt.ylabel('Count')
plt.xticks(rotation=25)
plt.show()
```

this plot shows the majority payment methods that a customer opts to pay with

```python
plt.hist(df['Review Rating'])
plt.title('Histogram of Review Rating')
plt.xlabel('Review Rating')
plt.ylabel('frequency')
plt.show()
```

by this plot we can see that, about 350 customers gave us 5.0 rating whereas about 650-700 customers gave us less then 3.0 rating 

```python
sns.countplot(x='Location', data=df, )
plt.title('Location')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

by this plot we can see that Montana, Callifornia, Illinois and Idano are the majority locations where customers have purchased items

```python
sns.countplot(x='Gender', data=df, )
plt.title('Gender')
plt.ylabel('Count')
plt.show()
```

though this plot, it is clearly visible that Male have purchased more(approx 2500-3000 items) than females(100-1500)

```python
sns.countplot(x='Season', data=df, )
plt.title('Season')
plt.ylabel('Count')
plt.show()
```

by this plot we can see the number of orders which takes placed in each of the seasons

```python
sns.countplot(x='Subscription Status', data=df, )
plt.title('Subscription Status')
plt.ylabel('Count')
plt.show()
```

we can see that most of the customers are not subscribed
